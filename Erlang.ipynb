```python
from scipy import integrate
import math as m
import pandas as pd
import numpy as np, matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
M_std_data = pd.read_csv('M_std_data.csv', sep=';',index_col=0)
```

```python
M_std_data
```

```python
def f(x,coef_before_x,coef_x,exp_coef):
    return coef_before_x*(x**coef_x)*m.exp(-exp_coef*x)
v, err = integrate.quad(f, start_a, end_b,args = (coef_before_x,coef_x,exp_coef))
print(v)
```

# решение примера

```python
M = 1  # 7 дней
M_real = 7
sigma = 4/M_real 
h = 1/M
k = round((1/sigma*h)**2)
k
```

```python
start_a = 3/7
end_b = 5/7
coef_before_x = k*h*((h*k)**(k-1))/m.factorial(k-1)
coef_x = (k-1)
exp_coef = h*k
coef_before_x
```

тут ответ более точный, так как в примере при вычислениях присутвуют округления

```python
v, err = integrate.quad(f, start_a, end_b,args = (coef_before_x,coef_x,exp_coef))
print(v)
```

```python
f2 = np.vectorize(f)
fig, ax = plt.subplots(figsize=(8, 3))
x = np.linspace(0, 4, 1000)
ax.plot(x, f2(x,coef_before_x,coef_x,exp_coef), lw=2)
ax.fill_between(x, f2(x,coef_before_x,coef_x,exp_coef), color='green', alpha=0.5)
ax.set_xlabel(" x", fontsize=18)
ax.set_ylabel(" f(x)", fontsize=18)
ax.set_ylim(0,3)
plt.show()
```
